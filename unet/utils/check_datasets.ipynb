# Dataset aligment consistency check
In this notebook we'll check if "egohands_dataset", "egtea_dataset" and "handsoverface_dataset" are consistent with the aligment. 

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
from tqdm.autonotebook import tqdm
import glob
import os

PROJECT_DIR = '/content/drive/MyDrive/Uni/CV Project/datasets/'
```

*Egohands*

```python
data_images = []
data_masks = []
data_boxes = []
dataset_name = 'egohands'

data_images = data_images + sorted(glob.glob(PROJECT_DIR + dataset_name + "/DATA_IMAGES/*"))
data_masks = data_masks + sorted(glob.glob(PROJECT_DIR + dataset_name + "/DATA_MASKS/*"))
data_boxes = data_boxes + sorted(glob.glob(PROJECT_DIR + dataset_name + "/DATA_BOXES/*"))

check = False
i = 0
for image in data_images:
  image_name = os.path.splitext(os.path.basename(image))[0]
  mask_name = os.path.splitext(os.path.basename(data_masks[i]))[0]
  box_name = os.path.splitext(os.path.basename(data_boxes[i]))[0]
  mask_name = mask_name.replace('Mask', 'Image')
  box_name = box_name.replace('Box', 'Image')
  i += 1
  if image_name != mask_name != box_name:
    check = True
    print(dataset_name + " is not aligned")
    print("First not aligned pair found: " + image_name + " / " + mask_name)
    break

if not check:
  print(dataset_name + " is ok")

print("Total number of usable images: ", len(data_images))
```
